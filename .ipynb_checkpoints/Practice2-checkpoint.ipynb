```python
%matplotlib inline

# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis,y_axis, 'o')
```

```python
plt.scatter(x_axis, y_axis, c='red', label='Chicago')
plt.ylabel('Savings ($)')
plt.xlabel('Month')
plt.title('Saving Account Results')
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()

ax.scatter(y_axis, x_axis, s=[i*5 for i in y_axis], c='skyblue', alpha = .8, edgecolor = 'k', linewidths = 2,label='Boston')
ax.set_title('PyBer Fare by Month')
ax.set_xlim(0,50)
ax.invert_yaxis()
ax.legend()
```

```python
plt.subplots(figsize=(8, 8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0,0,0,0,0,0,.2,0,0,0,0,0)
plt.pie(y_axis, labels=x_axis, explode = explode_values, autopct = '%.1f%%', colors=colors)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
explode = (0,0,.2,0,0,0,.2,0,0,0,0,0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y_axis,labels=x_axis, autopct = '%.1f%%', explode = explode, shadow = True, counterclock=False, startangle = 95, colors = colors)
plt.show()
```
